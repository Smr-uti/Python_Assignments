#### Q1.  Write a program that demonstrates all four LEGB levels in one nested structure:

```python

x = 'global'              # G
def outer():
    
    x = 'enclosing'       # E

    def inner():
        x = 'local'       # L
        print(x)          # x is from Local Scope
    inner()

outer()
print(x)                  # x is from Global scope
     
```

#### Q2. Repeat Q1 but remove x = 'local' from inner(). Now which x does inner() print? Explain the LEGB lookup path Python follows.


```python

x = 'global'              # G
def outer():
    x = 'enclosing'       # E

    def inner():
      # x = 'local'       # L
        print(x)          # x is from Enclosing Scope. Python searches x in local but did not find so it moved to enclosing where it found and print it.
    inner()

outer()
print(x)                  # x is from Global scope
     
```
