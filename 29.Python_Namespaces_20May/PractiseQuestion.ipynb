#### Q1.  Write a program that demonstrates all four LEGB levels in one nested structure:

```python

x = 'global'              # G
def outer():
    
    x = 'enclosing'       # E

    def inner():
        x = 'local'       # L
        print(x)          # x is from Local Scope. Local overlaps Enclosing and Global scopes.
    inner()

outer()
print(x)                  # x is from Global scope
     
```

#### Q2. Repeat Q1 but remove x = 'local' from inner(). 
#### Now which x does inner() print? 
#### Explain the LEGB lookup path Python follows.


```python

x = 'global'              # G
def outer():
    x = 'enclosing'       # E

    def inner():
      # x = 'local'       # L
        print(x)          # x is from Enclosing Scope. Python searches x in local but did not find so it moved to enclosing where it found and print it.
    inner()

outer()
print(x)                  # x is from Global scope
     
#LEGB Lookup Path:
## L - Local Scope - It defines variables inside a function. These variables accessible from only local scope.
## E - Enclosing Scope - It defines in Outer function and accessible from outer and inner function.
## G - Global Scope - It defines outside of all functions and accessible from anywhere in the code.
## B - Built-in Scope - It is the scope of built-in functions and keywords. These are accessible from anywhere.
```

#### Q11. Write the full payroll program from 10.py:
##### After running, annotate every variable inside compute_pay() with: L, E, G, or B.

```python
base_wage = 15.0              # Global Scope Variable
def payroll(hours):           # outer
    overtime_rate = 1.5       # Enclosing Scope Variable
    def compute_pay():        # inner
        regular  = min(hours, 40) * base_wage  # Local Scope Variable
        overtime = max(hours - 40, 0)          # Local Scope Variable
        ot_pay   = overtime * base_wage * overtime_rate    # Local Scope Variable   
        total    = regular + ot_pay        # Local Scope Variable
        print(f'Hours:    {hours}')        
        print(f'Regular:  ${regular:.2f}')
        print(f'OT Pay:   ${ot_pay:.2f}')
        print(f'Total:    ${total:.2f}')
    compute_pay()
payroll(50)
```

#### Q20. A student writes: What error occurs? What are the two ways to fix it? (Hint: one fix uses nonlocal)

```python
price = 100.0

def outer():
        price = 200.0          # enclosing
        def inner():
             global price
             price = price * 0.9   # tries to modify enclosing
             print(price)
        inner()
outer()
```

```python
price = 100.0

def outer():
    price = 200.0    
        
    def inner():
        nonlocal price
        price = price * 0.9   
        print(price)
    inner()
outer()
   
```
