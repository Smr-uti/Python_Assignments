#### Q1. What will be the output? What data type does input() always return?

#### Q3. What will be the output? What did int() do here?

```python
raw_value = input("Enter a number: ")
print(raw_value)
print(type(raw_value))
converted_value = int(raw_value) ## converting the string input into integer using int() function.
print(converted_value)
print(type(converted_value))
```

#### Q4. What will be the output? Explain Solution 1 -- convert after input.

```python
score1 = input("Enter first score: ")
score2 = input("Enter second score: ")
score1 = int(score1) ## converting the string input into integer using int() function.
score2 = int(score2)
total = score1 + score2
print(total)
```

#### Q6. What will be the output? What does id() return and what does it represent?


```python
roll = 10
print(roll)
print(id(roll)) ## id() function is used to get the memory address of the variable. It returns a unique identifier for the object.  
```

#### Q7. What will be the output? Does the id change when you reassign a variable to a new value?

```python
quantity = 10
print(quantity)
print(id(quantity)) 
quantity = 10.1
print(quantity)
print(id(quantity)) ## id is change here because string is immutable in python. We cant assign a new value to the same memory location.
```

#### Q8. What will be the output? What concept does this show about integer data type?

```python
item_count = 20
print(item_count)
print(id(item_count))
item_count = 30
print(item_count)
print(id(item_count)) ## here also integer is immutable in python. So when we assign a new value to the variable, it creates a new memory location for the new value and the id is change.
```

#### Q9. What will be the output? What is the difference between id of num1, num2, and result?

```python
val1 = input("Enter first value: ")
val2 = input("Enter second value: ")
print(id(val1))
print(id(val2))
joined = val1 + val2
print(joined)
print(id(joined)) ## id joined is different from id of val1 and val2.  
```

#### Q10. What will be the output? Does modifying a list element change its id?



```python
scores = [10, 20, 30]
print(scores)
print(id(scores))
scores[0] = 15
print(scores)
print(id(scores)) ## here the id of scores is same because list is mutable in python. We can change the value of the list without changing the memory location.
```

#### Q14. What will be the output? Why do id values of num1 and num2 differ even if values look the same?

```python
num1 = input("Enter first value: ")
print(id(num1))
num1 = int(num1)
print(type(num1))
num2 = input("Enter second value: ")
print(id(num2))
num2 = int(num2)
result = num1 + num2
print(result)
print(id(result)) ## the id of result is different from id of num1 and num2 because when we perform addition, it creates a new memory location for the new value and the id is change.  
```

#### Q15. What will be the output? Explain why integer is called immutable using this example.

```python
original = 10 ## integer is called immutable because once we assign a value to a variable, we cannot change the value of that variable. We can only reassign a new value to the variable, but the original value will remain unchanged. When we reassign a new value to the variable, it creates a new memory location for the new value and the id is change.  
print(original)
print(id(original))
original = 30
print(original)
print(id(original))
```
