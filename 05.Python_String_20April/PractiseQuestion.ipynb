```python
about_myself= """My name is Smruti
I am from Pune
I am from KRP College"""

print(about_myself)
```

# Slicing

#### 11. str6 = "welcome to india" — Slice and print only "welcome".

```python
str6="welcome to india"
print(str6[0:7])
```

#### 12.	From the same string, slice and print only "india".

```python
str6="welcome to india"
print(str6[10:16])
```

#### 13. From str6 = "welcome to india", use a step of 2 to print every alternate character from index 0 to 7.

```python
str6="welcome to india"
print(str6[0:16:2])
```

#### 14. str7 = "Artificial Intelligence" — Slice "Intelligence" from this string.

```python
str7="Artificial Intelligence"
print(str7[10:23])
```

#### 15. From str7 = "Artificial Intelligence", print the entire string using [::], then again using [0:23:1]. Confirm the outputs match.

```python
str7="Artificial Intelligence"
print(str7[::])

print(str7[0:23:1])
```
