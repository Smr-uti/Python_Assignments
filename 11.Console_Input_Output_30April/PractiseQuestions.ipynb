#### Q7. What will be the output? What does {} do inside format()?

```python
player = "Virat"
runs = 150
avg = 54.5
print("Player: {}, Runs: {}, Average: {}".format(player, runs, avg)) ## format() method, the {} curly braces act as placeholders or "slots" for the data you want to insert into the string.
```

#### Q8. What will be the output? Does the order of variables inside format() matter?

```python
brand = "Nike"
item = "Shoes"
cost = 4999
print("Brand: {} Item: {} Cost: {}".format(brand, item, cost)) ## Yes, the order of variables inside format() absolutely matters. 
##If you swap the order of the variables inside format(), the values will be plugged into the wrong placeholders, leading to 
##logical errors in your output.
```

#### Q14. What will be the output? What is the difference between normal print and string concatenation here?

```python
username = "Ganesh"
score = 88
score = str(score)
print("Normal:", "Username is", username, "Score is", score)
print("Concat:", "Username is " + username + " Score is " + score)
#Normal print and Concat print is give same output but to join variables in Normal we used (,) and in Concat we used (+)
```

#### Q15. What will be the output? Compare all three methods for the same data.


```python
dept = "HR"
emp = "Ramesh"
salary = 45000
print("Normal  :", "Dept:", dept, "Emp:", emp, "Salary:", salary)
print("format():", "Dept: {} Emp: {} Salary: {}".format(dept, emp, salary))
print(f"f-string : Dept: {dept} Emp: {emp} Salary: {salary}")

## Normal: using (,) print the data.
## format(): {} as a empty string at will be fill later by variables which are declare in .format().
## f-string: declare variable in {} for print the data.
```
