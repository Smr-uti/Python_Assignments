#### Q9. What will be the output? What does end="\n" do and is it different from default behavior?

```python
course = "Data Science"
duration = "6 months"
fees = 25000
print(course, end="\n") ##No, end="\n" is not different from the default behavior. By default, the print() function is programmed with end set to a newline character (\n).
print(duration, end="\n")
print(fees, end="\n")
print("Enrollment open")
```
