#### Q1. What will be the output? What is the difference between defining a function and calling a function?

```python
print("program started")

def display_info(): # what the function will do
    print("Welcome to Python class")
    print("Today topic is Functions")

display_info() # calling the function to execute the code inside the function

print("program execution completed")
```

#### Q2. What will be the output? What happens if you never call the function?

```python
print("Program Started")

def show_details():
    college = "MIT College"
    branch = "Computer Science"
    print("College:", college) # does not print because we are not calling the function
    print("Branch:", branch)  # does not print because we are not calling the function

print("Program execution completed")
```

#### Q3. What will be the output? How many times does the function body execute?

```python
print("Program Started")

def greet_user():
    print("Hello! Good Morning")
    print("Have a great day")

greet_user()
greet_user() # the function body execution 3 times because we are calling the function 3 times
greet_user()

print("Program Execution Completed")
```

#### Q4. What will be the output? What are x, y, z called inside the function definition?


```python
print("Program Started")

def cal_sum(x, y, z): # x, y, z parameters
    total = x + y + z
    print(total)

cal_sum(10, 20, 30)

print("Program Ends")
```

#### Q5. What will be the output? What are 100, 200, 300 called when passed to the function?


```python
print("Program Started")

def find_product(a, b, c):
    result = a * b * c
    print("Product:", result)

find_product(100, 200, 300) # positional arguments

print("Program Ends")
```
