#### Q9. What is the difference between these two functions? Which approach is better and why?

```python
def calc_with_print(a, b):
    print(a + b)

def calc_with_return(a, b): # this is better because we can use the return value in other calculations or print it later
    return a + b

calc_with_print(5, 10)

output = calc_with_return(5, 10)
print(output)
```

#### Q10. What will be the output if user enters 50 and 30? What does return allow us to do with the result?


```python
print("Program Started")

def find_max(num1, num2):
    if num1 > num2:
        return num1
    else:
        return num2

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
biggest = find_max(a, b) # the return value of the function is stored in the variable biggest
print("Biggest number:", biggest) # we can use the return value of the function to print it or use it in other calculations

print("Program Ends")
```

#### Q11. What will be the output if user enters 40 and 10? Can a function return more than one value?


```python
print("Program Starts")

def calc_add_sub(a, b):
    addition = a + b
    substraction = a - b
    return addition, substraction

num1 = int(input("Enter first No: "))
num2 = int(input("Enter second No: "))
res1, res2 = calc_add_sub(num1, num2) # the return values of the function are stored in the variables res1 and res2 

print("Result of Addition is:", res1)
print("Result of Substraction is:", res2)

print("Program Ends")
```

#### Q12. What will be the output if user enters 6 and 4? How many values does this function return?


```python
print("Program Started")

def all_operations(x, y):
    add = x + y
    sub = x - y
    mul = x * y
    return add, sub, mul 

v1 = int(input("Enter first No: "))
v2 = int(input("Enter second No: "))

r1, r2, r3 = all_operations(v1, v2) # function return 3 values

print("Addition: ", r1)
print("Substraction: ", r2)
print("Multiplication: ", r3)

print("Program Ends")
```

#### Q13. What will be the output? What is the function naming convention used in Python?


```python
def calculate_area(length, breadth): # snake_case function name
    area = length * breadth
    return area

def calculate_perimeter(length, breadth): # it should be small letter or seprate with underscore
    perimeter = 2 * (length + breadth)
    return perimeter

l = int(input("Enter length: "))
b = int(input("Enter breadth: "))

print("Area: ", calculate_area(l, b))
print("Perimeter: ", calculate_perimeter(l, b))

print("Program Ends")
```
