#### Q12. Write a function that returns multiple values: def student_stats(marks): Given marks = [85, 72, 91, 60, 78], return: highest, lowest, average Unpack and print all three.

```python
def student_stats(marks):
    average = sum(marks) / len(marks)

    highest = max(marks)
    lowest = min(marks)

    return highest, lowest, average

marks = [85, 72, 91, 60, 78]

high, low, avg = student_stats(marks)

print(f"Highest: {high}")
print(f"Lowest: {low}")
print(f"Average: {avg}")
```
