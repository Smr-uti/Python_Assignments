#### Q2. What will be the output? What is the difference between < and <= in while loop?

```python
counter = 0
while counter <= 5: # (<=) Less Than or Equal To: The loop continues while the condition is less than or equal to the target value.
    print("Counter:", counter)
    counter = counter + 1
print("Done")

# (<) Less Than: The loop continues only while the condition is strictly less than the target value.
```

#### Q11. What will be the output? What does break do here?

```python
for n in range(10):
    if n == 5: 
        break # it immediately terminates the entire for loop, regardless of how many iterations remain.
    print(n)
print("execution completed")
```

#### Q12. What will be the output? What is the difference between break and normal loop ending?

```python
for item in range(1, 8): # Normal loop: It ends naturally after processing every item in the range.
    if item == 4:
        break # It ends when condition is met.
    print("Item:", item)
print("Loop stopped")
```

#### Q13. What will be the output? What does continue do here?

```python
for n in range(10):
    if n == 5: 
        continue # The continue statement is used to skip the current iteration of a loop. Here skip the 5.
    print(n)
print("execution completed")
```

#### Q15. What will be the output? What is the difference between break and continue?


```python
for x in range(1, 6):
    if x == 3:
        break # When x reaches 3, the break command triggers. It immediately terminates the loop.
    print("break loop:", x)

for y in range(1, 6):
    if y == 3:
        continue # : When y reaches 3, the continue command triggers. This tells Python, Stop what you are doing for this specific number and jump back to the start of the loop for the next number.
    print("continue loop:", y)
```

#### Q19. What will be the output? When does a while loop become infinite?


```python
attempt = 1
while attempt <= 3: # loop's condition never becomes False
    print("Attempt:", attempt)
    attempt = attempt + 1
print("All attempts done")
```
