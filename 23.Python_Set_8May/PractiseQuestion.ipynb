#### Q12. What is the difference between aliasing and cloning a set? Write code to demonstrate both using id().

```python
set1={1,2,3,4,5}
set2=set1 # Aliasing= Multiple variables point to same memory location.
print(id(set1))
print(id(set2)) 
```

```python
set1={10,20,30,40,50}
set2=set1.copy() #  It create copy of set at different memory location.
print(id(set1))
print(id(set2))
```
