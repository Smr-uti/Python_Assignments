#### Q1. What will be the output if user enters 20? What will be the output if user enters 15?


```python
marks = int(input("Enter your marks: "))
if marks >= 35: ## 20 & 15 is less than 35 so skip the if block and excute next statement.
    print("You have passed the exam")
print("Result declared")

## Output: if user enters 20: Result declared
## if user enters 15: Result declared
```

#### Q2. What will be the output if user enters 500? What will be the output if user enters 200?


```python
balance = int(input("Enter your balance: "))
if balance >= 500: ## if enters 500. 500 is greater than equal to 500 this condition is true so Python execute if block.
    print("Transaction approved")
    print("Amount debited successfully")
print("Thank you for using our service") ## if user enters the 200 this condition is false so python skip the if block.

## Output: if user enter 500:
# Transaction approved
# Amount debited successfully
# Thank you for using our service

## Output: if user enter 200: 
# Thank you for using our service
```

#### Q13. What will be the output if speed=40? What if speed=70? What if speed=100?

#### Q15. What will be the output if username=admin and password=1234? What if username=admin and password=0000?
